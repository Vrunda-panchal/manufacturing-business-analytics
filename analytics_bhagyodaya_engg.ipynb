```python
import pandas as pd

vendors = pd.read_excel("vendors.xlsx")
customers = pd.read_excel("customers.xlsx")
purchases = pd.read_excel("purchases.xlsx")
sales = pd.read_excel("sales.xlsx")
highest_selling = pd.read_excel("highest_selling.xlsx")
```

```python
vendors.head()
```

```python
vendors.isnull().sum()
```

```python
vendors.dropna(inplace=True)
```

```python
vendors.drop_duplicates(inplace=True)
```

```python
vendors['Total Bills'] = pd.to_numeric(vendors['Total Bills'])
```

```python
vendors.sort_values(
    by='Total Bills',
    ascending=False
)
```

```python
customers.sort_values(
    by='Total',
    ascending=False
)
```

```python
sales.sort_values(
    by='Quantity Sold',
    ascending=False
)
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+mysqlconnector://root:Vrunda%40123@localhost/manufacturing"
)
```

```python
vendors.to_sql(
    'vendors',
    con=engine,
    if_exists='replace',
    index=False
)

customers.to_sql(
    'customers',
    con=engine,
    if_exists='replace',
    index=False
)

sales.to_sql(
    'sales',
    con=engine,
    if_exists='replace',
    index=False
)

purchases.to_sql(
    'purchases',
    con=engine,
    if_exists='replace',
    index=False
)
```
